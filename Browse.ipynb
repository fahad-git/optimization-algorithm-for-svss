```python
class Browse(QMainWindow):
   
    def __init__(self):
        super().__init__()
        self.force_quit = False
        self.__fading_button = None
        self.play_in_vlc_flag = True
        self.initUI()
        
    def initUI(self):
        
        QToolTip.setFont(QFont('SansSerif', 14))
        
        self.statusBar().showMessage('Ready')
        self.statusBar().setStyleSheet("color: white; text-align: center;font-size:20px;")
        
        self.resize(750, 600)
        self.center()
        self.setStyleSheet("background-color: #7585A0;")

            
        #         Layouts
        layout1 = QVBoxLayout()
        layout2 = QVBoxLayout()
        layout3 = QHBoxLayout() 

        lbl2 = QLabel('Browse', self)
        lbl2.resize(200,80)
        lbl2.move(275, 50)
        lbl2.setStyleSheet("color: white; text-align: center; font-size:30px;")
        lbl2.setAlignment(Qt.AlignCenter)
        
        layout1.addWidget(lbl2)

        
# Group                                                                          
                                                                                 
        disc_path = os.path.join("D:\\FYP\\description","description.xlsx")
        data = pd.ExcelFile(disc_path)
        df = data.parse('Sheet 1')
        df.set_index('Record',inplace=True)                                                                         

        mygroupbox = QGroupBox('List of Recorded videos')
        mygroupbox.setStyleSheet("color: white; text-align: center; font-size:16px;")
        grid = QGridLayout()


        style = "color: white; text-align: center;  font-size:14px;"
        
        lb = QLabel("<b>Record</b>")
        lb.setStyleSheet(style)
        lb.setAlignment(Qt.AlignCenter)
        grid.addWidget(lb, 0,0)
        
        lb = QLabel("<b>File Name</b>")
        lb.setStyleSheet(style)
        lb.setAlignment(Qt.AlignCenter)
        grid.addWidget(lb, 0,1)
        
        lb = QLabel("<b>Date</b>")
        lb.setStyleSheet(style)
        lb.setAlignment(Qt.AlignCenter)
        grid.addWidget(lb , 0,2)

        lb = QLabel("<b>Time</b>")
        lb.setStyleSheet(style)       
        lb.setAlignment(Qt.AlignCenter)
        grid.addWidget(lb , 0,3)
        
        lb = QLabel("<b>Duration</b>")
        lb.setStyleSheet(style)
        lb.setAlignment(Qt.AlignCenter)
        grid.addWidget(lb, 0,4)
        
        lb = QLabel("<b>Detections</b>")
        lb.setStyleSheet(style)
        lb.setAlignment(Qt.AlignCenter)
        grid.addWidget(lb , 0,5)
        
        lb = QLabel("<b>Status</b>")
        lb.setStyleSheet(style)
        lb.setAlignment(Qt.AlignCenter)
        grid.addWidget(lb , 0,6)
        
#         lb = QLabel('<b>Play Video</b>')
        lb = QLabel('')
        lb.setStyleSheet(style)
        grid.addWidget(lb, 0,7)

        mygroupbox.setLayout(grid)
        
        
        grid.addWidget(QLabel(""), 1,0)
        grid.addWidget(QLabel(""), 1,1)
        grid.addWidget(QLabel(""), 1,2)
        grid.addWidget(QLabel(""), 1,3)
        grid.addWidget(QLabel(""), 1,4)
        grid.addWidget(QLabel(""), 1,5)
        grid.addWidget(QLabel(""), 1,6)
        grid.addWidget(QLabel(""), 1,7)
        mygroupbox.setLayout(grid)
        
        self.labellist = []
        self.buttonlist1 = []
        i = 0
        for index, row in df.iterrows():
             # access data using column names
            obj = None
            for c , j in enumerate(eval(row['Detected_Objects'])):
                if obj == None:
                    obj = j + ", "
                    continue
                obj += j + ", "
                if c % 2 == 1:
                    obj += "\n"

            obj = obj[:-2]
            
            style = "color: white; text-align: center;font-size:12px;"
        
            lb = QLabel(str(index))
            lb.setStyleSheet(style)
            lb.setAlignment(Qt.AlignCenter)
            grid.addWidget(lb, i+2,0)
            
            lb = QLabel(str(row['Video_file_name']))
            self.labellist.append(lb)
            lb.setStyleSheet(style)
            lb.setAlignment(Qt.AlignCenter)
            grid.addWidget(lb, i+2,1)
            
            lb = QLabel(str(row['Date']))
            lb.setStyleSheet(style)
            lb.setAlignment(Qt.AlignCenter)
            grid.addWidget(lb , i+2,2)
            
            lb = QLabel(str(row['Time']))
            lb.setStyleSheet(style)
            lb.setAlignment(Qt.AlignCenter)
            grid.addWidget(lb , i+2,3)
            
            lb = QLabel(str(row['Duration']))
            lb.setStyleSheet(style)
            lb.setAlignment(Qt.AlignCenter)
            grid.addWidget(lb , i+2,4)
            
            lb = QLabel(str(obj))
            lb.setStyleSheet(style)
            lb.setAlignment(Qt.AlignCenter)
            grid.addWidget(lb , i+2,5)
            
            lb = QLabel(str(row['Status'])) 
            lb.setStyleSheet(style)
            lb.setAlignment(Qt.AlignCenter)
            grid.addWidget(lb, i+2,6)
            
            btn = QPushButton("Play")
            style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
            style += "color: white; padding: 5px 3px; text-align: center; text-decoration: none;" 
            style += " font-size:12px; }"
            btn.setStyleSheet(style)
            btn.clicked.connect(self.play_clicked)
            self.buttonlist1.append(btn)
            
            grid.addWidget(btn, i+2,7)
            
            mygroupbox.setLayout(grid)
            
            i += 1
            

        scroll = QScrollArea()
        scroll.setWidget(mygroupbox)
        scroll.setWidgetResizable(True)
        scroll.setFixedHeight(400)
        layout2.addWidget(scroll)
        
           
        #   Control Buttons
        player_btn = QPushButton("Video Player: VLC")
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += " font-size:12px; }"

        player_btn.setStyleSheet(style)
        player_btn.clicked.connect(self.player_setting)
        
        layout3.addWidget(QLabel(''))
        layout3.addWidget(player_btn)    
        layout3.addWidget(QLabel(''))
        
        back_btn = QPushButton("Back")
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += " font-size:12px; }"
        
        back_btn.setStyleSheet(style)
        back_btn.clicked.connect(self.go_back)  
        
        layout3.addWidget(back_btn)
        layout3.addWidget(QLabel(''))

        
        layout1.addLayout(layout2)
        layout1.addLayout(layout3)
        
#     Running controls ************************************ 
        widget = QWidget()
        widget.setLayout(layout1)
        self.setCentralWidget(widget)
        
        self.setWindowTitle('Smart Video Surveillance System')
        self.setWindowIcon(QIcon('icons/icon.png')) 
        self.show()

    @pyqtSlot() 
    def go_back(self):
        self.statusBar().showMessage("Back")
        self.cams = SVSS()
        self.cams.show()
        self.force_quit = True
        self.close()
        
    def player_setting(self):
        sender = self.sender()
        if self.play_in_vlc_flag:
            sender.setText("Video Player: Default")
            self.play_in_vlc_flag = False
        else:
            sender.setText("Video Player: VLC")
            self.play_in_vlc_flag = True
        
#         Play Videos
    def play_clicked(self):
        sender = self.sender()
        self.statusBar().showMessage('Playing Video')
        
        self.fade(sender)
        
        for i in range(len(self.buttonlist1)):
            if sender == self.buttonlist1[i]:
                video_name = self.labellist[i].text()
                
                video_path = os.path.join('D:\\FYP\\videos',video_name)
                
                if self.play_in_vlc_flag:
                    self.play_in_vlc(video_path)
                else:
                    player = VideoPlayer(video_path)
                    player.setWindowTitle("Player")
                    player.resize(600, 400)
                    player.show()
                                      
        self.statusBar().showMessage('Ready')
        

    def play_in_vlc(self,video_path):

        # For Async playing video
        process = subprocess.Popen(['vlc', video_path], 
                               stdout=subprocess.PIPE,
                               universal_newlines=True)
        
        
            # For Sync playing video
    
        while True:
            output = process.stdout.readline()
            return_code = process.poll()
            if return_code is not None:
                break
        
#     Click Animation
    def fade(self,btn):
        self.__fading_button = btn  # enter the "fading button" state
        self.__fading_button.setWindowOpacity(0.5)
        style = "QPushButton { border-radius: 10px; background-color: red; border: none;"
        style += "color: white; padding: 5px 3px; text-align: center; text-decoration: none;" 
        style += "font-size:12px; }"
        self.__fading_button.setStyleSheet(style)
        QTimer.singleShot(300, self.unfade)

    def unfade(self):
        self.__fading_button.setWindowOpacity(1)
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 5px 3px; text-align: center; text-decoration: none;" 
        style += " font-size:12px; }"
        self.__fading_button.setStyleSheet(style)
        self.__fading_button = None  # exit the "fading button" state


#         Center
    def center(self): 
        qr = self.frameGeometry()
        cp = QDesktopWidget().availableGeometry().center()
        qr.moveCenter(cp)
        self.move(qr.topLeft())

        
#   Close event
    def closeEvent(self, event):
        # Create a confirmation dialog asking for permission to quit the application:
        
        if self.force_quit:
            event.accept()
            return
        
        box = QMessageBox()
        box.setIcon(QMessageBox.Question)
        box.setWindowTitle("Message")
        box.setText("Are you sure to quit?")
        box.setStandardButtons(QMessageBox.Yes | QMessageBox.No)
        box.setDefaultButton(QMessageBox.No)
        box.setStyleSheet("background-color: #4E6EA5; color: white; text-align: center;  font-size:20px;")
        
        buttonYes = box.button(QMessageBox.Yes)
        buttonYes.setText("YES")
        buttonYes.setStyleSheet("background-color: #4CAF50; color: white; text-align: center;   font-size:16px;")

        buttonNo = box.button(QMessageBox.No)
        buttonNo.setText("NO")
        buttonNo.setStyleSheet("background-color: #4CAF50; color: white; text-align: center;  font-size:16px;")

        box.exec_()
        
        # Executing a routine for quitting the application:
        if box.clickedButton() == buttonYes:
            event.accept() 
        else:
            event.ignore()
        
```
