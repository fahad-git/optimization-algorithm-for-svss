```python
class Setup(QMainWindow):
   
    def __init__(self):
        super().__init__()
        self.force_quit = False
        self.initUI()
        
    def initUI(self):
        
        QToolTip.setFont(QFont('SansSerif', 16))
        
        self.statusBar().showMessage('Ready')
        self.statusBar().setStyleSheet("color: white; text-align: center;  font-size:12px;")
        
        self.resize(750, 600)
        self.center()
        self.setStyleSheet("background-color: #7585A0;")
        
        
# void QGridLayout::addWidget(QWidget *widget, int fromRow, int fromColumn, int rowSpan, 
# int columnSpan, Qt::Alignment alignment = Qt::Alignment())
# This is an overloaded function.        
        
#         Layouts
        layout1 = QVBoxLayout()
        layout2 = QVBoxLayout() 
        layout3 = QHBoxLayout() # layout 3 will be in layout 2
        layout4 = QGridLayout() # layout 4 will be in layout 3
        layout5 = QVBoxLayout() # layout 5 will be in layout 3
        

        lbl2 = QLabel('Setup', self)
        lbl2.resize(200,80)
        lbl2.move(275, 50)
        lbl2.setStyleSheet("color: white; text-align: center;  font-size:30px;")
        lbl2.setAlignment(Qt.AlignCenter)
        
        layout1.addWidget(lbl2)
        
#       Camera Source

        group_box2 =  QGroupBox('Others') # Letter group
        
        self.cam_radio_btn_group= QButtonGroup()
        cam1 =QRadioButton("Default Camera")
        cam2=QRadioButton("IP Camera")
        
        cam1.setStyleSheet("color: white; text-align: center; font-size:11px;") 
        cam2.setStyleSheet("color: white; text-align: center; font-size:11px;") 
        cam1.setChecked(True)
        cam1.toggled.connect(lambda:self.btnstate())
        cam2.toggled.connect(lambda:self.btnstate())
        
        self.cam_radio_btn_group.addButton(cam1,1)
        self.cam_radio_btn_group.addButton(cam2,2)
       
        

        
        layout = QGridLayout()
        layout.addWidget(cam1,0,0)
        layout.addWidget(cam2,1,0)


        
        self.lineEdit1 = QLineEdit("", self)
        self.lineEdit1.setGeometry(170, 200, 200, 30)
        self.lineEdit1.setStyleSheet("color: white; text-align: center; font-size:13px; padding: 5px 5px;")
        self.lineEdit1.setPlaceholderText("Disabled")
        self.lineEdit1.setReadOnly(True)
        
        layout.addWidget(self.lineEdit1,1,1,1,2)
        layout4.addLayout(layout,0,0,1,3)
        
#         Phone Number 

        lbl7 = QLabel('<b>Phone Number:</b>', self)
        lbl7.resize(130,30)
        lbl7.move(30, 400)
        lbl7.setStyleSheet("color: white; text-align: center;  font-size:13px;") 
        
        self.lineEdit4 = QLineEdit("", self)
        self.lineEdit4.setGeometry(170, 400, 200, 30)
        self.lineEdit4.setStyleSheet("color: white; text-align: center;  font-size:13px; padding: 5px 5px;")
        self.lineEdit4.setPlaceholderText("+92300000000")

        layout4.addWidget(lbl7,1,0)
        layout4.addWidget(self.lineEdit4,1,1,1,2)
        
        
#         Video Type
        
        lbl4 = QLabel('<b>Video Type:</b>', self)
        lbl4.resize(130,30)
        lbl4.move(30, 250)
        lbl4.setStyleSheet("color: white; text-align: center; font-size:13px;") 
        
        self.video_type_radio_group = QButtonGroup() # video type radio group
        va=QRadioButton("Auto")
        va.setChecked(True)
        vb=QRadioButton("avi")
        vc=QRadioButton("mp4")
        self.video_type_radio_group.addButton(va,3)
        self.video_type_radio_group.addButton(vb,4)
        self.video_type_radio_group.addButton(vc,5)
        
        grid = QGridLayout()
        grid.addWidget(va,0,0)
        grid.addWidget(vb,0,1)  
        grid.addWidget(vc,0,2)     

        
        layout4.addWidget(lbl4,2,0)
        layout4.addLayout(grid,2,1,1,2)

        
#        Resolutions

        lbl6 = QLabel('<b>Resolutions:</b>', self)
        lbl6.resize(130,30)
        lbl6.move(30, 350)
        lbl6.setStyleSheet("color: white; text-align: center; font-size:13px;") 
        
        self.resolution_radio_group = QButtonGroup() # Letter group
        ra=QRadioButton("480p")
        ra.setChecked(True)
        rb=QRadioButton("720p")
        rc=QRadioButton("1080p")
        
        self.resolution_radio_group.addButton(ra,6)
        self.resolution_radio_group.addButton(rb,7)
        self.resolution_radio_group.addButton(rb,8)
        
        
        grid = QGridLayout()
        grid.addWidget(ra,0,0)
        grid.addWidget(rb,0,1)  
        grid.addWidget(rc,0,2) 
  
        layout4.addWidget(lbl6,3,0)
        layout4.addLayout(grid,3,1,1,2) 
        
    
#         Frames per second

        lbl5 = QLabel('<b>Frames:</b>', self)
        lbl5.resize(130,30)
        lbl5.move(30, 300)
        lbl5.setStyleSheet("color: white; text-align: center; font-size:13px;") 

        
        self.lineEdit3 = QLineEdit("", self)
        self.lineEdit3.setGeometry(170, 300, 200, 30)
        self.lineEdit3.setStyleSheet("color: white; text-align: center; font-size:13px; padding: 5px 5px;")
        self.lineEdit3.setPlaceholderText("24")
        
        layout4.addWidget(lbl5,4,0)
        layout4.addWidget(self.lineEdit3,4,1,1,1)
        per_sec_lbl = QLabel('per second')
        per_sec_lbl.setStyleSheet("color: black; text-align: center; font-size:11px;") 
        layout4.addWidget(per_sec_lbl,4,2,1,1)
        
        group_box2.setLayout(layout4)
        
        newLayout = QGridLayout()
        newLayout.addWidget(group_box2,0,0,5,3)
        
#       Submit Button

        btn1 = QPushButton('Back', self)
        btn1.setToolTip('Click this to go to <b>home</b> ')  
        btn1.resize(100,100)
        btn1.clicked.connect(self.backbuttonClicked)
        btn1.setIcon(QIcon('icons/back_icon2.png'))
        btn1.setIconSize(QSize(40,40))
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += "font-size:20px; }"
        btn1.setStyleSheet(style)
        
        btn2 = QPushButton('Start', self)
        btn2.setToolTip('Click this to start <b>surveillance</b> ')  
        btn2.resize(100,100)
        btn2.clicked.connect(self.getDataOnSubmit)
        btn2.setIcon(QIcon('icons/surveillance_icon.png'))
        btn2.setIconSize(QSize(40,40))
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += " font-size:20px; }"
        btn2.setStyleSheet(style)
        
        submitButtonGrid = QGridLayout()
        submitButtonGrid.setSpacing(10)
        
        submitButtonGrid.addWidget(btn1,0,0)
        submitButtonGrid.addWidget(btn2,0,1)
        
        newLayout.addLayout(submitButtonGrid,5,0,1,3)
        
#  CheckBox

        mygroupbox = QGroupBox('List of Objects')
        grid = QGridLayout()
        
        self.labellist = []
        self.checkboxlist1 = []
        self.checkboxlist2 = []
        
#         Adding all option
        lbl_select_all = QLabel("Select")
        lbl_select_all.setStyleSheet("color: white; text-align: center;  font-size:12px;")
        self.labellist.append(lbl_select_all)
        
        record_main_check = QCheckBox("Record All")
        record_main_check.setStyleSheet("color: white; text-align: center;  font-size:12px;")
        record_main_check.toggled.connect(lambda:self.checkboxstate(record_main_check))
        self.checkboxlist1.append(record_main_check)
        
        notify_main_check = QCheckBox("Notify All")
        notify_main_check.setStyleSheet("color: white; text-align: center;  font-size:12px;")
        notify_main_check.toggled.connect(lambda:self.checkboxstate(notify_main_check))
        self.checkboxlist2.append(notify_main_check)
        
        grid.addWidget(self.labellist[0], 0,1)        
        grid.addWidget(self.checkboxlist1[0],0,3)
        grid.addWidget(self.checkboxlist2[0],0,4)
        
        
        for k,name in self.get_obj_list().items():
            entry = "{}{:<2} <b>{}</b>".format(k,'.',name)
            lbl_tmp = QLabel(entry)
            self.labellist.append(lbl_tmp)
            self.labellist[k].setAlignment(Qt.AlignLeft)  
            record_check = QCheckBox("Record")
            record_check.toggled.connect(lambda:self.checkboxstate(record_check))
            self.checkboxlist1.append(record_check)
            notify_check = QCheckBox("Notify")
            notify_check.toggled.connect(lambda:self.checkboxstate(notify_check))
            self.checkboxlist2.append(notify_check)

            
            grid.addWidget(self.labellist[k] , k,1)
            grid.addWidget(self.checkboxlist1[k],k,3)
            grid.addWidget(self.checkboxlist2[k],k,4)

        mygroupbox.setLayout(grid)
        
        scroll = QScrollArea()
        scroll.setWidget(mygroupbox)
        scroll.setWidgetResizable(True)
        scroll.setFixedHeight(400)
    
        layout5 = QVBoxLayout(self)
        layout5.addWidget(scroll)
        
#         Adding layouts
#         layout.setContentsMargins(left, top, right, bottom)
        
#       Layout5 is for Object list
        layout5.setContentsMargins(10, 0, 30, 0)
#       Layout4 is for text boxes
        layout4.setContentsMargins(10, 0, 20, 0)
        
        
        
        layout3.addLayout( layout5 )
        layout3.addLayout( newLayout  )
        layout2.addLayout( layout3 )
        layout1.addLayout( layout2 )
        
        
#    Frame Show

        widget = QWidget()
        widget.setLayout(layout1)
        self.setCentralWidget(widget)
        
        self.setWindowTitle('Smart Video Surveillance System')
        self.setWindowIcon(QIcon('icons/icon.png')) 
        self.show()

        
    def closeEvent(self, event):
        # Create a confirmation dialog asking for permission to quit the application:
        
        if self.force_quit:
            event.accept()
            return
        
        box = QMessageBox()
        box.setIcon(QMessageBox.Question)
        box.setWindowTitle("Message")
        box.setText("Are you sure to quit?")
        box.setStandardButtons(QMessageBox.Yes | QMessageBox.No)
        box.setDefaultButton(QMessageBox.No)
        box.setStyleSheet("background-color: #4E6EA5; color: white; text-align: center; font-size:20px;")
        
        buttonYes = box.button(QMessageBox.Yes)
        buttonYes.setText("YES")
        buttonYes.setStyleSheet("background-color: #4CAF50; color: white; text-align: center; font-size:16px;")

        buttonNo = box.button(QMessageBox.No)
        buttonNo.setText("NO")
        buttonNo.setStyleSheet("background-color: #4CAF50; color: white; text-align: center; font-size:16px;")

        box.exec_()

        # Executing a routine for quitting the application:
        if box.clickedButton() == buttonYes:
            event.accept() 
        else:
            event.ignore() 


    def center(self): 
        qr = self.frameGeometry()
        cp = QDesktopWidget().availableGeometry().center()
        qr.moveCenter(cp)
        self.move(qr.topLeft())

#         ============================= Radio Button Click =============================
        
    def btnstate(self):
        b = self.cam_radio_btn_group.checkedButton() 
        if b.text() == "Default Camera":
            self.lineEdit1.setText('')
            if b.isChecked() == True:
                self.lineEdit1.setPlaceholderText("Disabled")
                self.lineEdit1.setReadOnly(True)
            else:
                self.lineEdit1.setReadOnly(False)
                self.lineEdit1.setPlaceholderText("Enter IP Address or URL")
                
        elif b.text() == "IP Camera":
            self.lineEdit1.setText('')
            if b.isChecked() == True:
                self.lineEdit1.setReadOnly(False)
                self.lineEdit1.setPlaceholderText("Enter IP Address or URL")
            else:
                self.lineEdit1.setPlaceholderText("Disabled")
                self.lineEdit1.setReadOnly(True)


#         ============================= Checkbox Click =============================
                
    def checkboxstate(self,cbox):
        if cbox.text() == "Record All":
            if cbox.isChecked():
                for i in range(len(self.checkboxlist1)):
                    self.checkboxlist1[i].setChecked(True)
            else:
                for i in range(len(self.checkboxlist1)):
                    self.checkboxlist1[i].setChecked(False)
        elif cbox.text() == "Notify All":
            if cbox.isChecked():
                for i in range(len(self.checkboxlist1)):
                    self.checkboxlist1[i].setChecked(True)
                    self.checkboxlist2[i].setChecked(True)
            else:
                for i in range(len(self.checkboxlist1)):
                    self.checkboxlist1[i].setChecked(False)
                    self.checkboxlist2[i].setChecked(False)
        elif cbox.text() == "Record":
            if not cbox.isChecked():
                for i in range(len(self.checkboxlist1)):
                    if not self.checkboxlist1[i].isChecked():
                        self.checkboxlist2[i].setChecked(False)
        
        else:
            for i in range(len(self.checkboxlist2)):
                if self.checkboxlist2[i].isChecked():
                    self.checkboxlist1[i].setChecked(True)
            
        
 
    @pyqtSlot() 
    def backbuttonClicked(self):
        self.statusBar().showMessage("Back")
        self.cams = SVSS()
        self.cams.show()
        self.force_quit = True
        self.close()
        
    @pyqtSlot()
    def getDataOnSubmit(self):
        
        source_camera = self.cam_radio_btn_group.checkedButton().text()
        source_camera_path = None
        
        if source_camera == "IP Camera":
            source_camera_path = self.lineEdit1.text()
    
        else:
            source_camera_path = 0
        
        phone_number = self.lineEdit4.text()
        
        ext = self.video_type_radio_group.checkedButton().text()
        if ext == "Auto":
            ext = "avi"
        
        video_file_name = "video."+ext
        
        resolutions = self.resolution_radio_group.checkedButton().text()
        
        frames_per_second = self.lineEdit3.text() 
        
        
        list_of_obj = []
        list_of_notify = []
    
        for name,rec,notify in zip(self.labellist,self.checkboxlist1,self.checkboxlist2):
            if name.text() == "Select":
                continue
            name=name.text().split('.')[1]
            TAG_RE = re.compile(r'<[^>]+>')
            name = TAG_RE.sub('', name)
            name = name.strip()
            
            if notify.isChecked():
                list_of_obj.append(name)
                list_of_notify.append(name)
            elif rec.isChecked():
                list_of_obj.append(name)
          
        data = (source_camera_path,phone_number,video_file_name,resolutions,frames_per_second,list_of_obj,list_of_notify)
        
        self.cams = Surveillance(data)
        self.cams.show()
        self.force_quit = True
        self.close()
        
        
    def get_obj_list(self):
        obj_list = {
            1:"person",
            2:"bicycle",
            3:"car",
            4:"motorcycle",
            5: "airplane",
            6: "bus",
            7: "train",
            8:"truck",
            9: "boat",
            10:"bird",
            11:"cat",
            12:"dog",
            13:"horse",
            14:"sheep",
            15:"cow",
            16:"elephant",
            17:"bear",
            18:"zebra",
            19:"giraffe",
            20:"suitcase",
            21:"fork",
            22:"knife",
            23:"laptop",
            24:"mouse",
            25:"scissors",
            26:"cell phone",
            27:"baseball bat",
            28:"tennis racket"
        }

        return obj_list
```
