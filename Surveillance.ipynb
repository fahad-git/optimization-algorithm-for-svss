```python
class Surveillance(QMainWindow):
   
    def __init__(self,data):
        super().__init__()
        self.force_quit = False
        self.stop_surveillance = False
        self.start_time = None
        self.start_time_flag = False
        
        self.base_path = data[0]
        self.phone_number = data[1]
        self.video_file_name = data[2]
        self.resolutions = data[3]
        self.frames_per_second = int(data[4])
        self.list_of_detections = data[5]
        self.list_of_notify_obj = data[6]
        
        self.STD_DIMENSIONS =  {
            "480p": (640, 480),
            "720p": (1280, 720),
            "1080p": (1920, 1080),
            "4k": (3840, 2160),
        }

        # This dictionary return video writer object according to video type
        self.VIDEO_TYPE = {
            'avi': cv.VideoWriter_fourcc(*'XVID'),
            'mp4': cv.VideoWriter_fourcc(*'XVID'),
        }
        
        self.initUI()
        
    def initUI(self):
        
        QToolTip.setFont(QFont('SansSerif', 14))
        
        self.statusBar().showMessage('Ready')
        self.statusBar().setStyleSheet("color: white; text-align: center;  font-size:20px;")
        
        self.resize(750, 600)
        self.center()
        self.setStyleSheet("background-color: #7585A0;")
        
        
        
        #         Layouts
        layout1 = QVBoxLayout()
        layout2 = QVBoxLayout() 
        layout3 = QHBoxLayout() # layout 3 will be in layout 2
        layout4 = QGridLayout() # layout 4 will be in layout 3
        layout5 = QVBoxLayout() # layout 5 will be in layout 3
        

        lbl2 = QLabel('SURVEILLANCE', self)
        lbl2.resize(200,80)
        lbl2.move(275, 50)
        lbl2.setStyleSheet("color: white; text-align: center;   font-size:30px;")
        lbl2.setAlignment(Qt.AlignCenter)
        
        layout1.addWidget(lbl2)

#  Cam Show Setup

        self.scene = QGraphicsScene()
        self.view = QGraphicsView(self.scene)
        self.view.resize(640, 480);
        self.view.setFixedSize( 400, 400 );
        
        layout5.addWidget(self.view)
        
        # this is for timer
        layout = QGridLayout()
        
        self.lbl15 = QLabel('Data', self)
        self.lbl15.resize(200,80)
        self.lbl15.move(275, 50)
        self.lbl15.setStyleSheet("color: white; text-align: center;  font-size:11px;")
        
        self.lbl16 = QLabel('Time', self)
        self.lbl16.resize(200,80)
        self.lbl16.move(275, 50)
        self.lbl16.setStyleSheet("color: white; text-align: center;   font-size:11px;")
        
        self.lbl17 = QLabel('Surveillance Time', self)
        self.lbl17.resize(200,80)
        self.lbl17.move(275, 50)
        self.lbl17.setStyleSheet("color: white; text-align: center;  font-size:11px;")
        
        layout.addWidget(self.lbl15,0,0)
        layout.addWidget(self.lbl16,0,1)
        layout.addWidget(self.lbl17,0,2) 
        
        layout5.addLayout(layout)
        
# ============================== Message Labels Start ================================================
        
#      Video File Name (dynamic)

        lbl3 = QLabel('Video File Name: ', self)
        lbl3.resize(200,80)
        lbl3.move(275, 50)
        lbl3.setStyleSheet("color: white; text-align: center; font-size:11px;")
        # lbl4 is for Video File Name
        self.lbl4 = QLabel('', self)
        self.lbl4.resize(200,80)
        self.lbl4.move(275, 50)
        self.lbl4.setStyleSheet("color: white; text-align: center; font-size:11px;")
        self.lbl4.setText('')
        
        layout4.addWidget(lbl3,0,0)
        layout4.addWidget(self.lbl4,0,1)

#      Resolutions (static)

        lbl7 = QLabel('Camera Resolutions: ', self)
        lbl7.resize(200,80)
        lbl7.move(275, 50)
        lbl7.setStyleSheet("color: white; text-align: center; font-size:11px;")
        
        lbl8 = QLabel('', self)
        lbl8.resize(200,80)
        lbl8.move(275, 50)
        lbl8.setStyleSheet("color: white; text-align: center;  font-size:11px;")
        lbl8.setText(self.resolutions)

        layout4.addWidget(lbl7,1,0)
        layout4.addWidget(lbl8,1,1)

#      Frames per second (static)
        
        lbl9 = QLabel('Frames per Second: ', self)
        lbl9.resize(200,80)
        lbl9.move(275, 50)
        lbl9.setStyleSheet("color: white; text-align: center;   font-size:11px;")
        
        lbl10 = QLabel('', self)
        lbl10.resize(200,80)
        lbl10.move(275, 50)
        lbl10.setStyleSheet("color: white; text-align: center;   font-size:11px;")
        lbl10.setText(str(self.frames_per_second))
        
        layout4.addWidget(lbl9,2,0)
        layout4.addWidget(lbl10,2,1)

#      Detected Objects (dynamic)
        
        lbl11 = QLabel('Detections: ', self)
        lbl11.resize(200,80)
        lbl11.move(275, 50)
        lbl11.setStyleSheet("color: white; text-align: center; font-size:11px;")
        
        self.lbl12 = QLabel('', self)
        self.lbl12.resize(200,80)
        self.lbl12.move(275, 50)
        self.lbl12.setStyleSheet("color: white; text-align: center;  font-size:11px;")
        
        layout4.addWidget(lbl11,3,0)
        layout4.addWidget(self.lbl12,3,1) 
        
#      Notification Message (dynamic)
        
        lbl13 = QLabel('Notification: ', self)
        lbl13.resize(200,80)
        lbl13.move(275, 50)
        lbl13.setStyleSheet("color: white; text-align: center;  font-size:11px;")
        
        self.lbl14 = QLabel('', self)
        self.lbl14.resize(200,80)
        self.lbl14.move(275, 50)
        self.lbl14.setStyleSheet("color: white; text-align: center;   font-size:11px;")
        
        layout4.addWidget(lbl13,4,0)
        layout4.addWidget(self.lbl14,4,1)     
    

# ============================== Message Labels End ================================================

        
#       Submit Button

        self.btn1 = QPushButton('Back', self)
        self.btn1.setToolTip('Click this to go to <b>home</b> ')  
        self.btn1.resize(100,100)
        self.btn1.clicked.connect(self.stop_video_surveillance)
        self.btn1.setIcon(QIcon('icons/stop_icon.png'))
        self.btn1.setIconSize(QSize(40,40))
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += "font-size:20px; }"
        self.btn1.setStyleSheet(style)
        
        btn2 = QPushButton('Start', self)
        btn2.setToolTip('Click this to start <b>surveillance</b> ')  
        btn2.resize(100,100)
        btn2.clicked.connect(self.start_Surveillance)
        btn2.setIcon(QIcon('icons/start_icon.png'))
        btn2.setIconSize(QSize(40,40))
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += " font-size:20px; }"
        btn2.setStyleSheet(style)
        
        layout4.addWidget(self.btn1,5,0)
        layout4.addWidget(btn2,5,1)
        
#         Browse and Home Button
        self.btn3 = QPushButton('Browse', self)
        self.btn3.setToolTip('Click this to go to <b>home</b> ')  
        self.btn3.resize(100,100)
        self.btn3.clicked.connect(self.go_to)
        self.btn3.setIcon(QIcon('icons/browse.png'))
        self.btn3.setIconSize(QSize(40,40))
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += " font-size:20px; }"
        self.btn3.setStyleSheet(style)
        
        btn4 = QPushButton('Home', self)
        btn4.setToolTip('Click this to start <b>surveillance</b> ')  
        btn4.resize(100,100)
        btn4.clicked.connect(self.go_to)
        btn4.setIcon(QIcon('icons/home.png'))
        btn4.setIconSize(QSize(40,40))
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += " font-size:20px; }"
        btn4.setStyleSheet(style)
        
        layout4.addWidget(self.btn3,6,0)
        layout4.addWidget(btn4,6,1)

#         Browse and Home Button ============

#         Adding layouts
#       layout.setContentsMargins(left, top, right, bottom)

#       Layout 4 is for Labeling (Grid Layout)
#       Layout 5 is for showing frame 
        layout5.setContentsMargins(5, 10, 20, 0)
        layout4.setContentsMargins(5, 10, 10, 0)
        
        layout3.addLayout( layout5 )
        layout3.addLayout( layout4 )
        layout2.addLayout( layout3 )
        layout1.addLayout( layout2 )
#        
        
#    Frame Show

        widget = QWidget()
        widget.setLayout(layout1)
        self.setCentralWidget(widget)
        
        self.setWindowTitle('Smart Video Surveillance System')
        self.setWindowIcon(QIcon('icons/icon.png')) 
        
        
        img = Image.open('icons/Cam.png')
        enhancer = ImageEnhance.Brightness(img)
        img = enhancer.enhance(2)
        img = img.resize((640, 480))
        self.display_image(img)
        
#         Timer code will go here

        self.timer = QTimer(self)
        self.timer.timeout.connect(self.showTime)
        self.timer.start(1000) # update every second
        self.showTime()
        
        self.show()
        
        
        

    def start_Surveillance(self):
        self.fade(self.sender())
        self.statusBar().showMessage('Surveillance')
        self.btn1.setText("Stop")
        now = datetime.now()
        time = now.strftime("%H:%M:%S")
        self.start_time = time
        self.start_time_flag = True
        txt = "{:^17}\n{:^17}".format("Surveillance Time","00:00:00")
        self.lbl17.setText(txt)
        
        self.start()
#         Here will be code for multi-threading

        self.stop_surveillance = False
        img = Image.open('icons/Cam.png')
        enhancer = ImageEnhance.Brightness(img)
        img = enhancer.enhance(2)
        img = img.resize((640, 480))
        self.display_image(img)
        
        
    def stop_video_surveillance(self):
        self.fade(self.sender())  
        text = self.sender().text()
        if text == "Stop":        
            self.statusBar().showMessage('Surveillance Stopped!')
            self.start_time_flag = False
            self.stop_surveillance = True
            self.sender().setText("Back")
        
        else:
            self.timer.stop()
            self.statusBar().showMessage("Back")
            self.set_duration()
            self.cams = Setup()
            self.cams.show()
            self.force_quit = True
            self.close()
        
    def go_to(self):
        sender = self.sender()
        self.fade(sender)
        self.timer.stop()
        self.set_duration() 
        
        if sender == self.btn3:
            self.statusBar().showMessage("Going to Browse...")
            self.cams = Browse()
            self.cams.show()
            self.force_quit = True
            self.close()
        else:
            self.statusBar().showMessage("Going Home...")
            self.cams = SVSS()
            self.cams.show()
            self.force_quit = True
            self.close()
        
        
    def display_image(self, img):
        self.scene.clear()
        w, h = img.size
        self.imgQ = ImageQt.ImageQt(img)  # we need to hold reference to imgQ, or it will crash
        pixMap = QPixmap.fromImage(self.imgQ)
        self.scene.addPixmap(pixMap)
        self.view.fitInView(QRectF(0, 0, w, h), Qt.KeepAspectRatio)
        self.scene.update()            

#     Click Animation
    def fade(self,btn):
        self.__fading_button = btn  # enter the "fading button" state
        self.__fading_button.setWindowOpacity(0.5)
        style = "QPushButton { border-radius: 10px; background-color: red; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += " font-size:20px; }"
        self.__fading_button.setStyleSheet(style)
        QTimer.singleShot(200, self.unfade)

    def unfade(self):
        self.__fading_button.setWindowOpacity(1)
        style = "QPushButton { border-radius: 10px; background-color: #4CAF50; border: none;"
        style += "color: white; padding: 10px 10px; text-align: center; text-decoration: none;" 
        style += "font-size:20px; }"
        self.__fading_button.setStyleSheet(style)
        self.__fading_button = None  # exit the "fading button" state        
 


    def showTime(self):
        currentTime = QTime.currentTime()
        displayTxt = currentTime.toString('hh:mm:ss')
        
        now = datetime.now()
        date = now.strftime("%d-%b-%Y")
        time = now.strftime("%H:%M:%S")
        
        txt = "{:^17}\n{:^17}".format("Date",date)
        self.lbl15.setText(txt)
        
        txt = "{:^17}\n{:^17}".format("Time",displayTxt)
        self.lbl16.setText(txt)
        
        if self.start_time_flag:
            t1 = datetime.strptime(self.start_time,"%H:%M:%S")
            t2 = datetime.strptime(time,"%H:%M:%S")
            diff = t2 - t1
            txt = "{:^17}\n{:^17}".format("Surveillance Time",str(diff))
            self.lbl17.setText(txt)
        
               
        
# ============================ Start Method  ===================================================================

## This method will start the system. This method takes 3 parameters. Video file name, frames per second and 
## resolutions to record video
    def start(self):
        # this line will start video camera with the help of library OpenCV
        camera = cv.VideoCapture(0)

        # In this line video dimensions are set. 
        dims = self.get_dims(camera,res=self.resolutions)
        # In this line video type is defined
        video_type_cv = self.get_video_type(self.video_file_name)
        ## Here starts the logic for classification at runtime
        previous_frame = None
        frame_no = 0
        video_no = len(os.listdir(video_dir))+1
        while True:
            try:
                if self.stop_surveillance:
                    break
                
                captured, current_frame = camera.read()
                # If this is true means frame is captured
                if captured == True:
                    current_frame = current_frame[:,::-1]
                    
#                   Showing image
                    self.statusBar().showMessage('Surveillance...')
                    cv.imwrite("tmp/img1.jpg",current_frame)
                    img = Image.open('tmp/img1.jpg')
                    enhancer = ImageEnhance.Brightness(img)
                    img = enhancer.enhance(1)
                    QCoreApplication.processEvents()
                    self.display_image(img)
#                  ****************************************

                    result = self.are_different(previous_frame,current_frame)

                    # If frames are different then start classifying
                    if result == True:
                        name = "classified-{}.jpg".format(frame_no)
                        frame_no += 1
                        detections = self.detectObjects(current_frame,name)
                        video_name = "video{}.avi".format(video_no)
                        self.lbl4.setText(video_name)
                        # Here is detection method which will take decision of comparing object and setuptext file
                        record_video = self.create_description(detections,video_name)

                        if record_video == True:        
                            video_no += 1
                            self.video_file_name = os.path.join(video_dir,video_name)
                            out = cv.VideoWriter(self.video_file_name,video_type_cv,self.frames_per_second,dims)                        
                            out.write(current_frame)
                            # If Object is detected then record next 200 frames.
                            for i in range(144):
                                if self.stop_surveillance:
                                    camera.release()
                                    out.release()
                                    return;
                                cap, frame = camera.read()
                                if cap == True:
                                    frame = frame[:,::-1]
                #                   Showing image
                                    self.statusBar().showMessage('Recording...')
                                    cv.imwrite("tmp/img1.jpg",frame)
                                    img = Image.open('tmp/img1.jpg')
                                    enhancer = ImageEnhance.Brightness(img)
                                    img = enhancer.enhance(1)
                                    QCoreApplication.processEvents()
                                    self.display_image(img)
                #                  ****************************************
                                    out.write(frame)
                            out.release()

                previous_frame = current_frame

            except Exception as e: 
                print(e)
                print("Terminated...!")
                break

        camera.release()
        print("Recording ended", end="")
        print("!!!")


# ============================ Start Method End ===================================================================
# ============================ Get Dims Method Start ==============================================================
# Standard Video Dimensions Sizes

    def change_res(self,cap, width, height):
        cap.set(3, width)
        cap.set(4, height)

    # grab resolution dimensions and set video capture to it. By default resolutions are 480p
    def get_dims(self,cap, res='480p'):
        if res in self.STD_DIMENSIONS:
            width,height = self.STD_DIMENSIONS[res]
        ## change the current caputre device
        ## to the resulting resolution
        self.change_res(cap, width, height)
        return width, height

    # This method returns video type. It first split file name and return video type according to video extension.
    def get_video_type(self,filename):
        filename, ext = os.path.splitext(filename)
        if ext in self.VIDEO_TYPE:
            return  self.VIDEO_TYPE[ext]
        return self.VIDEO_TYPE['avi']
# ============================ Get Dims Method End =================================================================

# ============================ Frame Differencing Method Start ======================================================

        # This method will return boolean based on frame difference
    # To calculate differences we have used means squared error (MSE)
    def are_different(self,x1,x2):
        try:
            x1 = cv.cvtColor(x1, cv.COLOR_BGR2GRAY)
            x2 = cv.cvtColor(x2, cv.COLOR_BGR2GRAY)

            err = np.sum((x1.astype("float") - x2.astype("float")) ** 2)
            err /= float(x1.shape[0] * x2.shape[1])

            if err < 100:
                return False
        except:
            print("Starting Surveillance...")

        return True

# ============================ Frame Differencing Method Start ======================================================

# ============================ Detect Object Method Start ===========================================================

    def detectObjects(self,image,name):
    #     detections = detector.detectObjectsFromImage(image, output_image_path=os.path.join(output_dir , "classified-" + img_name),
    #                                                  minimum_percentage_probability=30)

        custom_objects = detector.CustomObjects(person = True,  bicycle = True,  car= True, motorcycle= True,
                                                airplane= True, bus= True, train= True,truck= True,boat= True,
                                                bird= True,   cat= True,   dog= True,   horse= True,   sheep= True,
                                                cow= True,elephant= True,bear = True,zebra= True,giraffe= True,
                                                suitcase= True,fork= True,knife= True,laptop= True,   mouse= True,
                                                scissors= True,cell_phone = True,baseball_bat=True, tennis_racket=True)

        detections = detector.detectCustomObjectsFromImage(
                                            custom_objects=custom_objects,
    #     detections = detector.detectObjectsFromImage(
                                            input_type="array",
                                            input_image = image, 
                                            output_image_path=os.path.join(output_dir ,name),
                                            extract_detected_objects=False,
                                            minimum_percentage_probability=50,
                                            display_percentage_probability=True,
                                            display_object_name=True
                                            )

        return detections

# ============================ Detect Object Method End ===========================================================

# ============================ Create Description Method Start =====================================================

    def write_into_file(self,PATH,values):

        try:

            df2  = pd.DataFrame([values], columns=['Record','Video_file_name','Detected_Objects','Date','Time','Duration','Status'] )
            df2.set_index('Record',inplace=True)

            data = pd.ExcelFile(PATH)
            df = data.parse('Sheet 1')
            df.set_index('Record',inplace=True)

            df = pd.concat([df,df2])

            with pd.ExcelWriter(PATH) as writer:
                    df.to_excel(writer, sheet_name='Sheet 1')

        except:
            with pd.ExcelWriter(PATH) as writer:
                df2.to_excel(writer, sheet_name='Sheet 1')

    def create_description(self,detections,video_name):
        disc_file_name= "description.xlsx"
        disc_path = os.path.join(description_dir,disc_file_name)

        detect = []
        label_text = ""
        k = 0
        for eachobj in detections:
            if eachobj["name"] in self.list_of_detections:
                detect.append(eachobj["name"])
                label_text += ", "+eachobj["name"]
                k += 1
                if k % 3 == 0:
                    label_text += "\n"

        self.lbl12.setText(label_text[1:])
        if detect == []:
            return False

        rec_no = 'row'+video_name.split(".")[0][5:]
        # datetime object containing current date and time
        now = datetime.now()
        date = now.strftime("%d-%b-%Y")
        time = now.strftime("%H:%M:%S")

        lst = None

        if list(set(detect) & set(self.list_of_notify_obj)) == []:
            lst = [rec_no,video_name,detect,date,time,None,"Record"]
        else:
            lst = [rec_no,video_name,detect,date,time,None,"Notify"]
            msg = detect[0]
            for i in detect[1:]:
                msg += ", " + i
            msg += " detected in the camera!"
            self.notify(msg)

    #   lst = ['row3','test2.avi','Person,Cat','1/3/2019','11:01 AM','30 sec','Nil']
        self.write_into_file(disc_path,lst)
        return True
# ============================ Create Description Method End ===========================================================
# ============================ Duration Method Start ===========================================================
    def get_duration(self,file_path):
        clip = VideoFileClip(file_path)
        dur = int(clip.duration)
        hrs, mins, secs = dur//60//60, dur//60%60, dur%60
        hrs = "0"+str(hrs) if(hrs<10) else str(hrs)
        mins = "0"+str(mins) if(mins<10) else str(mins)
        secs = "0"+str(secs) if(secs<10) else str(secs)
    #     print("Length of the video: "+ hrs +":"+ mins +":"+ secs)
        clip.close()
        return hrs +":"+ mins +":"+ secs


    def set_duration(self):

        disc_path = os.path.join(description_dir,"description.xlsx")

        data = pd.ExcelFile(disc_path)
        df = data.parse('Sheet 1')
        df.set_index('Record',inplace=True)
        indices = df[pd.isnull(df['Duration'])].index
        df['Duration'] = df['Duration'].astype('str')
        for index in indices:
            vid = df.loc[index,'Video_file_name']
            video_path = os.path.join(video_dir,vid)
            duration = self.get_duration(video_path)
            df.at[index,'Duration'] = duration

        with pd.ExcelWriter(disc_path) as writer:
            df.to_excel(writer, sheet_name='Sheet 1')


# ============================ Duration Method End =============================================================

# ============================ Duration Method End =============================================================

    def notify(self,msg):

        # Your Account SID from twilio.com/console
        account_sid = "AC498981900ebe3f343b680c4b3ca1d45a"
        # Your Auth Token from twilio.com/console
        auth_token  = "3743284c476eaf8ffe4ac24559e7c211"

        client = Client(account_sid, auth_token)

        message = client.messages.create(
            to=self.phone_number, 
            from_="+18302713672",
            body=msg)
        
        msg1 = msg[:-23]        
        msg2 = msg[-23:]
        
        str_msg1 = "Notify against id: " + str(message.sid)
        str_msg2 = "Message:\n'" + msg1 + "'\n"+msg2+"\nSent to: " + self.phone_number
        
        self.lbl14.setText(str_msg2)
        self.statusBar().showMessage(str_msg1)

# ============================ Duration Method End =============================================================
    
# ==================================================================================================================

    def center(self): 
        qr = self.frameGeometry()
        cp = QDesktopWidget().availableGeometry().center()
        qr.moveCenter(cp)
        self.move(qr.topLeft())
           
    def closeEvent(self, event):
        # Create a confirmation dialog asking for permission to quit the application:
        
        if self.force_quit:
            event.accept()
            return
        
        box = QMessageBox()
        box.setIcon(QMessageBox.Question)
        box.setWindowTitle("Message")
        box.setText("Are you sure to quit?")
        box.setStandardButtons(QMessageBox.Yes | QMessageBox.No)
        box.setDefaultButton(QMessageBox.No)
        box.setStyleSheet("background-color: #4E6EA5; color: white; text-align: center;   font-size:20px;")
        
        buttonYes = box.button(QMessageBox.Yes)
        buttonYes.setText("YES")
        buttonYes.setStyleSheet("background-color: #4CAF50; color: white; text-align: center;  font-size:16px;")

        buttonNo = box.button(QMessageBox.No)
        buttonNo.setText("NO")
        buttonNo.setStyleSheet("background-color: #4CAF50; color: white; text-align: center;   font-size:16px;")

        box.exec_()

        # Executing a routine for quitting the application:
        if box.clickedButton() == buttonYes:
            event.accept() 
        else:
            event.ignore() 
```
